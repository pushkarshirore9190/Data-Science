```python
bank_products_categories = ['Savings Account', 'Current Account', 'Credit Card', 'Loan', 'Insurance']
```

```python
def generate_bank_products(num_products):
    return ', '.join(sample(bank_products_categories, num_products))
```

```python
import pandas as pd
import numpy as np

# Load the existing CSV file
# existing_df = pd.read_csv('your_existing_file.csv')

# Number of new rows to add
num_new_rows = 400

# Generate random data for new rows with conditions
np.random.seed(42)  # For reproducibility

new_data = {
    'Age': np.random.randint(18, 65, size=num_new_rows),
    'Annual Income': np.random.randint(30000, 1000000, size=num_new_rows),
    'Occupation': np.random.choice(['Salaried', 'Housewife', 'Business', 'Other'], size=num_new_rows),
    'AccountBalance': np.random.randint(10000, 500000, size=num_new_rows),
    'Credit Score(in range of 300,900)': np.random.randint(300, 901, size=num_new_rows),
    'Gender': np.random.choice(['Male', 'Female'], size=num_new_rows),
    'Education': np.random.choice(['Graduate', 'Under Graduate', 'Post Graduate'], size=num_new_rows),
    'Total Active loans': np.random.randint(0, 5, size=num_new_rows),
    'Bank Products': [generate_bank_products(np.random.randint(1, 4)) for _ in range(num_new_rows)],  # 1 to 3 products
    'Fixed Deposit': np.random.choice(['Yes', 'No'], size=num_new_rows),  # Corrected syntax
    'Tenure': np.random.randint(1, 20, size=num_new_rows),
    'Transaction in last Month': np.random.randint(10, 100, size=num_new_rows),
    'Networth': np.random.randint(10000, 5000000, size=num_new_rows),  # Corrected syntax
    'Married': np.random.choice(['yes', 'no'], size=num_new_rows),
    'Dependents': np.random.randint(0, 5, size=num_new_rows),
    'Vehicle': np.random.choice(['yes', 'no'], size=num_new_rows),
    'House': np.random.choice(['own', 'rented'], size=num_new_rows)
}

# Convert to DataFrame
new_df = pd.DataFrame(new_data)

# Apply conditions: If Age < 24, then Married = 'no' and Networth < 30000
new_df.loc[new_df['Age'] < 24, 'Married'] = 'no'
new_df.loc[(new_df['Age'] < 24) & (new_df['Networth'] > 30000), 'Networth'] = np.random.randint(10000, 30000)

# Optionally, add some cases where Networth < 30000 for ages >= 24
indices = new_df.loc[new_df['Age'] >= 24].sample(frac=0.1).index
new_df.loc[indices, 'Networth'] = np.random.randint(10000, 30000, size=len(indices))

# Load the existing DataFrame (assuming it's loaded earlier)
# existing_df = pd.read_csv('your_existing_file.csv')

# Concatenate the new rows to the existing DataFrame
```
