```python
from sklearn.cluster import DBSCAN
from sklearn.ensemble import IsolationForest
from sklearn.covariance import EllipticEnvelope
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import numpy as np
```

```python
x,y = make_blobs(n_samples=500, centers=3, cluster_std=0.4,random_state=0)
```

```python
plt.scatter(x[:,0], x[:,1])
```

```python
db = DBSCAN(eps=0.3, min_samples=10).fit(x)
```

```python
plt.scatter(x[:,0], x[:,1],c=y)
```

```python
plt.scatter(x[:,0], x[:,1],c=db.labels_)
```

```python
x,y = make_blobs(n_samples=2000, centers=2, cluster_std=0.9,random_state=0)
plt.scatter(x[:,0], x[:,1])
```

```python
ee = EllipticEnvelope(contamination=0.1)
ee.fit(x)
```

```python
prediction = ee.predict(x)
```

```python
plt.scatter(x[:,0], x[:,1],c=prediction)
```

```python
iso = IsolationForest()
```

```python
iso.fit(x)
```

```python
pred = iso.predict(x)
```

```python
plt.scatter(x[:,0], x[:,1],c=pred)
```

```python
from sklearn.neighbors import LocalOutlierFactor
```

```python
clf = LocalOutlierFactor(n_neighbors=3)
clf.fit(x)
```

```python
pred = clf.fit_predict(x)
```

```python
plt.scatter(x[:,0], x[:,1],c=pred)
```

```python
x = np.array([[1,5],[1,6],[1,7],[2,5],[2,6],[6,5],[9,1]])
```

```python
plt.scatter(x[:,0],x[:,1])
plt.show()
```

```python
X = x[:,0]
y = x[:,1]

X = X.reshape(-1,1)
clf = LocalOutlierFactor(n_neighbors=3)
clf.fit(X)
```

```python
pred = clf.fit_predict(X)
plt.scatter(X, y,c=pred)
```
