```python
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_regression
```

```python
x,y = make_regression(n_features=1, noise=5, n_samples=10000,random_state=0)
```

```python
x
```

```python
y
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
