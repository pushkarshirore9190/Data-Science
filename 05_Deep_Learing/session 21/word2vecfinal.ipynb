```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```

```python
import pandas as pd
```

```python
datasett = pd.read_csv('/Users/shridharmankar/review.csv')
```

```python
datasett
```

```python
from nltk.corpus import stopwords
import re
corpus = []
for i in range(0, 29):
    review = re.sub('[^a-zA-Z]', ' ', datasett['review'][i])
    review = review.lower()
    review = review.split() 
    review = [lemmatizer.lemmatize(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess

sentences = []
for sent in corpus:
    sent_token=sent_tokenize(sent)
    for sent in sent_token:
        sentences.append(simple_preprocess(sent))
```

```python
sentences
```

```python
import gensim
model=gensim.models.Word2Vec(sentences,window=3,min_count=1)
```

```python
model.wv.index_to_key
```

```python
from gensim.models import KeyedVectors
word_vectors = model.wv
word_vectors.save("word2vec.wordvectors")

wv = KeyedVectors.load("word2vec.wordvectors")
vector = wv['worst']
```

```python
import numpy as np
def sentence_vector(sent):
    doc = [word for word in sent.split() if word in model.wv.index_to_key]
    return np.mean(model.wv[doc], axis=0)
```

```python
sentence_vector(datasett['review'].values[20])
```

```python
from tqdm import tqdm
```

```python
final = []
for s in tqdm(datasett['review'].values):
    final.append(sentence_vector(s))
```
